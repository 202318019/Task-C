```python
name = [span.text.strip() for span in soup.find_all('span', class_='a-size-medium a-color-base a-text-normal')]
print(name)
```

```python
price = [tag.text.strip() for tag in value]
print(price)
```

```python
date = [tag.text.strip() for tag in deliv]
print(date)
```
